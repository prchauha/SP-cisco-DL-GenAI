----------------------------------
#### box plots for numeric data from the California Housing dataset 
- and interpretations
--------------------------------

```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load California Housing dataset
california_housing = fetch_california_housing()
data = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
```

```python
data.sample(5)
```

```python
# Select numeric columns for box plots
numeric_columns1 = ['MedInc', 
                    'HouseAge', 
                    'AveRooms', 
                    'AveBedrms']

numeric_columns2 = ['Population', 
                    'AveOccup', 
                    'Latitude', 
                    'Longitude']
```

```python
# Create box plots for each numeric column
plt.figure(figsize=(15, 4))
for i, column in enumerate(numeric_columns1, 1):
    plt.subplot(1, 4, i)
    plt.boxplot(data[column])
    plt.title(column)

plt.tight_layout()
```

**MedInc (Median Income):**

- `Symmetry`: It appears to have a rightward skew, indicating that the distribution is skewed towards higher median incomes.
- `Skewness`: Rightward skew.
- `Outliers`: Some potential outliers on the higher-income side.

**HouseAge (House Age):**

- Symmetry: Appears relatively symmetric.
- Skewness: Minimal skewness.
- Outliers: No obvious outliers.

**AveRooms (Average Rooms):**

- Symmetry: Rightward skew, suggesting more variability in houses with a higher number of rooms.
- Skewness: Rightward skew.
- Outliers: Potential outliers on the higher side.

**AveBedrms (Average Bedrooms):**

- Symmetry: Rightward skew, indicating more variability in houses with a higher number of bedrooms.
- Skewness: Rightward skew.
- Outliers: Potential outliers on the higher side.

```python
# Create box plots for each numeric column
plt.figure(figsize=(15, 4))
for i, column in enumerate(numeric_columns2, 1):
    plt.subplot(1, 4, i)
    plt.boxplot(data[column])
    plt.title(column)

plt.tight_layout()
```
