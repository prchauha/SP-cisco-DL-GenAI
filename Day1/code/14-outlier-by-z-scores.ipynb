```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.stats import zscore
```

```python
# Load the Iris dataset
iris = load_iris()
data = iris.data
```

```python
# Assume two outliers in the sepal length (the first column)
# Replace two values with extreme values to simulate outliers
data[0, 0] = 8.0
data[1, 0] = 8.5
```

```python
# Calculate Z-scores for sepal length
z_scores_sepal_length = zscore(data[:, 0])
```

```python
# Set a Z-score threshold (e.g., |Z| > 2 is commonly used)
threshold = 2
```

```python
# Identify outliers based on the threshold
outliers = np.where(np.abs(z_scores_sepal_length) > threshold)[0]
```

```python
# Plot the Z-scores with outliers highlighted
plt.figure(figsize=(8, 6))
plt.scatter(range(len(data)), z_scores_sepal_length, color='blue', label='Z-scores (Sepal Length)')
plt.scatter(outliers, z_scores_sepal_length[outliers], color='red', label='Outliers')
plt.axhline(y=threshold, color='orange', linestyle='--', label='Threshold')
plt.axhline(y=-threshold, color='orange', linestyle='--')
plt.title('Z-score Plot with Outliers Highlighted (Sepal Length)')
plt.xlabel('Data Point Index')
plt.ylabel('Z-score (Sepal Length)')
plt.legend();
```

#### display the rows which are outliers 

```python
import pandas as pd
```

```python
# Display the rows corresponding to outliers
outlier_rows = pd.DataFrame(data[outliers], columns=iris.feature_names)
print("Rows corresponding to outliers:")
print(outlier_rows)
```
