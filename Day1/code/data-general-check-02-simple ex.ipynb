```python
# Import necessary libraries
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
```

**iris data**

```python
# Load the Iris dataset from scikit-learn
iris = load_iris()
```

```python
# Create a Pandas DataFrame from the dataset
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=columns)
```

```python
iris_df.sample(10)
```

```python
# Summary statistics of numerical features
print("\nSummary Statistics:")
print(iris_df.describe())
```

```python
# Check for any missing values
print("\nMissing Values:")
print(iris_df.isnull().sum())
```

```python
# Check for duplicate rows
print("\nDuplicate Rows:")
print(iris_df.duplicated().sum())
```

```python
# Explore unique values in the target variable
print("\nUnique Values in the Target Variable:")
print(iris_df['target'].value_counts())
```

```python
# Visualize the distribution of numerical features
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(iris_df, hue='target')
```

```python
# Visualize basic statistics with a boxplot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df.drop('target', axis=1))
plt.title("Boxplot of Features")
plt.show()
```

**breast cancer data**

```python
from sklearn.datasets import load_breast_cancer
```

```python
# Load the Breast Cancer dataset
cancer = load_breast_cancer()
data = np.c_[cancer.data, cancer.target]
columns = np.append(cancer.feature_names, ['target'])
```

```python
# Create a Pandas DataFrame
df_cancer = pd.DataFrame(data, columns=columns)
```

```python
df_cancer.sample(5)
```

```python
df_cancer.info()
```

```python
df_cancer.columns
```

```python
df_cancer.describe()
```

```python
print("\nUnique Values in Target:")
print(df_cancer['target'].value_counts())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df_cancer.isnull().sum())
```

```python
# Check for duplicate rows
duplicate_rows_cancer = df_cancer[df_cancer.duplicated()]
print("\nDuplicate Rows:")
print(duplicate_rows_cancer)
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df_cancer.drop('target', axis=1))
plt.title("Boxplot of Features in Breast Cancer Dataset")
plt.show()
```

```python
# Visualize box plots for each feature
plt.figure(figsize=(18, 14))
for i, feature in enumerate(cancer.feature_names):
    plt.subplot(5, 6, i + 1)
    sns.boxplot(x='target', y=feature, data=df_cancer, showfliers=False)
    plt.title(f"Boxplot of {feature}")

plt.tight_layout()
plt.show()
```

```python
# Visualize KDE plots for each feature
plt.figure(figsize=(18, 14))
for i, feature in enumerate(cancer.feature_names):
    plt.subplot(5, 6, i + 1)
    sns.kdeplot(df_cancer[feature][df_cancer['target'] == 0], label='Malignant', shade=True)
    sns.kdeplot(df_cancer[feature][df_cancer['target'] == 1], label='Benign', shade=True)
    plt.title(f"KDE Plot of {feature}")

plt.tight_layout()
plt.show()
```

#### Data Quality Checks (Sanity) for a Dataset

1. **Missing Values Check**
   - Identify and quantify missing values in each column.
   - Check for patterns or correlations in missing data.

2. **Duplicate Entries**
   - Identify and remove duplicate rows.
   - Ensure uniqueness of key columns.

3. **Data Type Validation**
   - Verify that each column has the correct data type (e.g., numeric, categorical, date).
   - Check for inconsistencies in data types.

4. **Range and Boundaries**
   - Check if numerical values fall within expected ranges.
   - Validate that dates and times are within reasonable boundaries.

5. **Value Uniqueness**
   - Ensure that categorical variables contain only expected values.
   - Check for unexpected or out-of-scope values.

6. **Consistency Checks**
   - Verify consistency across related columns (e.g., start and end dates).
   - Ensure that calculated fields (e.g., totals, ratios) are correct.

8. **Range Validations for Categorical Data**
   - Ensure categorical variables do not have unexpected values or typos.
   - Validate categories against a predefined list of valid values.

10. **Format Validation**
    - Ensure that string fields conform to expected formats (e.g., email addresses, phone numbers).
    - Validate that date fields follow the correct date format.

11. **Temporal Consistency**
    - Verify that temporal data (e.g., timestamps, durations) follows logical sequences.
    - Check for future or unrealistic dates.

13. **Data Sampling**
    - Perform random sampling to manually check for data accuracy and completeness.
    - Validate that samples are representative of the full dataset.

14. **Statistical Summaries**
    - Review basic statistical summaries (mean, median, standard deviation) to spot anomalies.
    - Compare summaries across different segments or groups.

15. **Data Normalization**
    - Check if numerical data is scaled or normalized correctly.
    - Verify that data transformations are applied consistently.

16. **Historical Comparison**
    - Compare current data with historical data to identify discrepancies or unexpected changes.
    - Assess trends and patterns over time.

17. **Data Consistency with External Sources**
    - Validate data against external sources or benchmarks.
    - Check for alignment with known standards or reference datasets.
