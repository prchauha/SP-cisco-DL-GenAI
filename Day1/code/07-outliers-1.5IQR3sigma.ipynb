```python
import numpy as np
import matplotlib.pyplot as plt

# Create some example data
np.random.seed(42)
data = np.random.normal(0, 1, 1000)  # Sample data from a normal distribution

# Calculate quartiles and IQR
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1

# Calculate the 1.5 IQR limits
lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr

# Calculate the 3 sigma limits
std_dev = np.std(data)
lower_limit_sigma = np.mean(data) - 3 * std_dev
upper_limit_sigma = np.mean(data) + 3 * std_dev

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 8))

# Plot the data in the first subplot
ax1.hist(data, bins=50, color='blue', alpha=0.7)
ax1.axvline(lower_limit, color='red', linestyle='dashed', linewidth=2, label='1.5 IQR Lower Limit')
ax1.axvline(upper_limit, color='red', linestyle='dashed', linewidth=2, label='1.5 IQR Upper Limit')
ax1.legend()
ax1.set_title('1.5 IQR Rule')

# Plot the data in the second subplot
ax2.hist(data, bins=50, color='green', alpha=0.7)
ax2.axvline(lower_limit_sigma, color='purple', linestyle='dashed', linewidth=2, label='3 Sigma Lower Limit')
ax2.axvline(upper_limit_sigma, color='purple', linestyle='dashed', linewidth=2, label='3 Sigma Upper Limit')
ax2.legend()
ax2.set_title('3 Sigma Rule')

# Show the plots
plt.tight_layout()
plt.show()
```
