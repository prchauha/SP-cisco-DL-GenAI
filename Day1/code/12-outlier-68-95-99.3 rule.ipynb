```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data from a normal distribution
np.random.seed(42)
data = np.random.normal(0, 1, 1000)
```

```python
# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)
```

```python
# Create a histogram
plt.figure(figsize=(15, 6))
plt.hist(data, bins=30, density=True, color='lightgray', edgecolor='black')

# Plot lines and annotate the 68-95-99.7 rule
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=2, label='+1 Std Dev')
plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=2, label='-1 Std Dev')
plt.axvline(mean + 2 * std_dev, color='orange', linestyle='dashed', linewidth=2, label='+2 Std Dev')
plt.axvline(mean - 2 * std_dev, color='orange', linestyle='dashed', linewidth=2, label='-2 Std Dev')
plt.axvline(mean + 3 * std_dev, color='purple', linestyle='dashed', linewidth=2, label='+3 Std Dev')
plt.axvline(mean - 3 * std_dev, color='purple', linestyle='dashed', linewidth=2, label='-3 Std Dev')

# Annotate the plot with percentages
plt.annotate('68%', xy=(mean + 0.1, 0.25), color='green', fontsize=12)
plt.annotate('95%', xy=(mean + 1.5, 0.25), color='orange', fontsize=12)
plt.annotate('99.7%', xy=(mean + 2.7, 0.25), color='purple', fontsize=12)

# Set plot labels and title
plt.title('68-95-99.7 Rule: Empirical Rule for a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show legend
plt.legend();
```
